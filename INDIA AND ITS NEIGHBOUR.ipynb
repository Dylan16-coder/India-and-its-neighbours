# INDIA AND ITS NEIGHBOURS

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
india_stat=pd.read_excel('India Statistics.xlsx')
china_stat=pd.read_excel('China Statistics.xlsx')
pakistan_stat=pd.read_excel('Pakistan Statistics.xlsx')
srilanka_stat=pd.read_excel('Sri Lanka Statistics.xlsx')
bhutan_stat=pd.read_excel('Bhutan Statistics.xlsx')
nepal_stat=pd.read_excel('Nepal Statistics.xlsx')
bangladesh_stat=pd.read_excel('Bangladesh_Statistics.xlsx')
allcountries_stat=pd.read_excel('all_countries.xlsx')
```

#   India Statistics

```python
# 'ip' stands for India Population
# 'ig' stands for India GDP
# 'igg' stands for India GDP Growth
sns.set_theme(style="darkgrid")
ip=sns.catplot(x="Year",y="Population", data=india_stat,palette="flare",kind="bar",height=10,aspect=1)
ip.set_xlabels('Year',fontsize=15)
ip.set_ylabels('Population',fontsize=15)


sns.set_theme(style="darkgrid")
ig=sns.catplot(x="Year",y="GDP", data=india_stat,palette="flare",kind="bar",height=10,aspect=1)
ig.set_xlabels('Year',fontsize=15)
ig.set_ylabels('GDP(in billions)',fontsize=15)

sns.set_theme(style="darkgrid")
igg=sns.catplot(x="Year",y="gdpgrowth", data=india_stat,palette="flare",kind="bar",height=10,aspect=1)
igg.map(plt.axhline, y=0, color='dimgrey', linewidth=2)
igg.set_xlabels('Year', fontsize=15) 
igg.set_ylabels('GDP Growth(in percentage)',fontsize=15)
```

# INDIA AND CHINA

# Population

```python
w=0.4
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(8)

bar1=np.arange(len(india_stat.Population))
bar2=[i+w for i in bar1]

plt.bar(bar1,india_stat.Population,w,color='yellowgreen',label="India")
plt.bar(bar2,china_stat.Population,w,color='darkcyan',label="China")


plt.xlabel("Year",size=15)
plt.ylabel("Population",size=15)
plt.title("India vs China(Population)",size=15)
plt.xticks(bar1+w/2,india_stat.Year)
plt.legend()
plt.show()

f = plt.figure()
f.set_figwidth(15)
f.set_figheight(8)
plt.plot(india_stat.Year,india_stat.Population,color='yellowgreen',linewidth=3,marker=".",markersize=15,markeredgecolor='dimgrey')
plt.plot(india_stat.Year,china_stat.Population,color='darkcyan',linewidth=3,marker=".",markersize=15,markeredgecolor='dimgrey')
plt.legend(['India',"China"])
plt.xlabel('Year',size=15)
plt.ylabel('Population',size=15)
plt.title("India vs China(Population)",size=15)

plt.show()

```
